### 4. Services & Subscription Behavior

Customer subscription choices can significantly influence churn behavior. Value-added services such as technical support, security features, and backup services may increase customer satisfaction and reduce the likelihood of switching providers.

This section examines the relationship between service subscriptions and customer churn, with the goal of identifying services that contribute to stronger customer retention.

##### a) Internet Service Type vs Customer Churn

Goal:
* Understand how different internet service types relate to customer churn.
* This helps determine whether certain service offerings are associated with higher churn risk.

The output from Tableau is presented below:

<img src="../Visualisations/Internet Service Type vs Customer Churn.png" width="200" height="600" />

**Key insights**:

| Internet Service    | Churn Rate |
| ------------------- | ---------- |
| DSL                 | **18.96%** |
| Fiber Optic         | **41.89%** |
| No Internet Service | **7.40%**  |


1. Fiber Optic Customers Have the Highest Churn

Customers using fiber optic internet have a churn rate of 41.9%, which is more than double that of DSL customers and nearly six times higher than customers without internet service.

This suggests that fiber optic customers represent the highest-risk customer segment.

Possible reasons include:

* Higher service pricing

* Greater competition in high-speed internet markets

* Higher expectations for performance and reliability

2. DSL Customers Show Moderate Churn

DSL users have a churn rate of 18.9%, which is significantly lower than fiber optic but still notable.

This could indicate that DSL customers are relatively stable but may still be influenced by pricing or service quality.

3. Customers Without Internet Service Rarely Churn

Customers without internet service have a churn rate of 7.4%, the lowest among all groups.

This likely reflects:

* Lower monthly costs

* Simpler service packages

* Fewer competitive alternatives



**Business Implications**

The company should prioritize customer retention strategies for fiber optic users, since they represent the most volatile segment.

Potential strategies include:

* Improving fiber service reliability

* Offering bundled services (tech support, security, streaming)

* Providing loyalty discounts for long-term fiber customers

# Conclusion: Customer Churn Analysis

The exploratory data analysis (EDA) of the telecommunications dataset provides several **clear insights into customer churn behavior**, highlighting patterns related to **contracts, tenure, revenue, and service subscriptions**.


### 1. Overall Churn Overview

* Out of 7,043 customers, **1,869 churned**, giving an overall churn rate of **26.5%**.
* Churn is concentrated among **shorter-tenure customers** and those with **lower total charges**, indicating that **newer customers are the highest-risk group**.



### 2. Contract & Tenure Analysis

* **Month-to-month contracts** have the highest churn rate (**42.7%**), while **one-year** and **two-year contracts** have significantly lower churn (**11.3% and 2.8%**, respectively).
* Longer contracts provide **stability and reduced churn**, emphasizing the importance of incentivizing customers toward longer-term plans.


### 3. Revenue & Charges Analysis

* **Monthly Charges:** Customers with higher monthly bills generally exhibit higher churn, though the trend is non-linear.
* **Total Charges:** Customers with **low total charges (<500)** have the highest churn (**41.5%**), while high-spending long-term customers show very low churn (**2.8–12%**).
* These findings confirm that **early-stage customers and lower tenure users are at the highest risk of leaving**.


### 4. Services & Subscription Behavior

* **Internet Service:** Fiber optic users churn the most (**41.9%**) compared to DSL (**18.9%**) and no internet (**7.4%**).
* **Tech Support:** Customers without tech support churn at **41.6%**, while those with support churn only **15.2%**.
* **Online Security:** Lack of security services is associated with high churn (**41.8%**) versus **14.6%** for subscribers.
* **Streaming Services:** Minor effect; streaming TV subscribers churn slightly less (**30.1%**) than non-subscribers (**33.5%**).

**Key Insight:** Value-added services like Tech Support and Online Security are **the strongest retention drivers**, while entertainment services have only a modest effect.


### 5. Actionable Recommendations

1. **Focus on Early-Stage Customers:** Introduce onboarding, engagement, and loyalty programs to reduce churn among new customers.
2. **Promote Long-Term Contracts:** Encourage month-to-month customers to switch to one- or two-year contracts through incentives.
3. **Highlight Value-Added Services:** Offer Tech Support and Online Security bundles, especially to high-risk fiber-optic and new customers.
4. **Monitor High-Paying Customers:** Address service quality concerns for high-monthly-bill or fiber-optic users to prevent preventable churn.


### Overall Summary

The analysis reveals that **customer churn is heavily influenced by contract type, tenure, and value-added service subscriptions**. Early intervention strategies targeting **new and high-risk customers**, combined with **promoting support and security services**, can significantly reduce churn and increase long-term revenue.

This EDA provides a **strong foundation for predictive modeling** to identify at-risk customers and implement **targeted retention strategies**.
